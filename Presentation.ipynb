```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the finalized experimental results
# (These results are automatically generated by the project's run_experiments.py pipeline)
results_df = pd.read_csv('final_results.csv')

# Format the dataframe for presentation
pd.set_option('display.float_format', '{:.2f}'.format)
print("================ Evaluation Metrics Summary ================")
display(results_df)
```

```python
# Plotting the Accuracies
plt.figure(figsize=(12, 7))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x='Model', y='Accuracy (%)', data=results_df, palette='viridis')

plt.title('Tracking Accuracy Comparison: Federated TCN vs Baselines', fontsize=16, fontweight='bold')
plt.ylabel('Tracking Accuracy (%)', fontsize=14)
plt.xlabel('Model Architecture', fontsize=14)

# Zoom in the Y-axis to clearly highlight the significant >5% accuracy gap
plt.ylim(80, 100)
plt.xticks(rotation=15, fontsize=12)
plt.yticks(fontsize=12)

# Add explicit value labels on top of the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f') + '%',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', 
                xytext=(0, 10), 
                textcoords='offset points',
                fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```
